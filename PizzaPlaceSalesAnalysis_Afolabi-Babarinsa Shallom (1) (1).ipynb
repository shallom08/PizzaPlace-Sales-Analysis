```python
# Import required libraries
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# Set up visualization preferences
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['font.size'] = 10
```

```python
print("Starting Pizza Place Sales Analysis...")
print("="*50)

# Load all CSV files
print("Loading data files...")

from IPython.display import display

orders = pd.read_csv(r"C:\Users\CYBER SHAY\Downloads\Pizza+Place+Sales\pizza_sales\orders.csv")
order_details = pd.read_csv(r"C:\Users\CYBER SHAY\Downloads\Pizza+Place+Sales\pizza_sales\order_details.csv") 
pizzas = pd.read_csv(r"C:\Users\CYBER SHAY\Downloads\Pizza+Place+Sales\pizza_sales\pizzas.csv")
pizza_types = pd.read_csv(r"C:\Users\CYBER SHAY\Downloads\Pizza+Place+Sales\pizza_sales\pizza_types.csv", encoding="latin1")

# Display all dataframes
display(orders)
display(order_details)
display(pizzas)
display(pizza_types)
```

```python
print("\n=== DATASET OVERVIEW ===\n")

print(f"\033[1m{'Orders':<15}\033[0m {orders.shape[0]:,} records, {orders.shape[1]} columns")
print(f"\033[1m{'Order Details':<15}\033[0m {order_details.shape[0]:,} records, {order_details.shape[1]} columns")
print(f"\033[1m{'Pizzas':<15}\033[0m {pizzas.shape[0]:,} records, {pizzas.shape[1]} columns")
print(f"\033[1m{'Pizza Types':<15}\033[0m {pizza_types.shape[0]:,} records, {pizza_types.shape[1]} columns")
```

```python
# Data Quality Check
print("\nData Quality Assessment:")
print("-" * 30)

datasets = {'Orders': orders, 'Order Details': order_details, 
           'Pizzas': pizzas, 'Pizza Types': pizza_types}

for name, df in datasets.items():
    missing = df.isnull().sum().sum()
    duplicates = df.duplicated().sum()
    print(f"{name}: {missing} missing values, {duplicates} duplicates")
```

```python
# Merge orders with order_details (using order_id)
df = orders.merge(order_details, on="order_id", how="inner")

# Merge with pizzas (using pizza_id)
df = df.merge(pizzas, on="pizza_id", how="inner")

# Merge with pizza_types (using pizza_type_id)
df = df.merge(pizza_types, on="pizza_type_id", how="inner")

# Final dataset
print("=== FINAL DATASET ===")
print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")

df.head()
```

### [6] Peak hours of sales?

```python
print("\nPEAK HOURS ANALYSIS")
print("=" * 50)

# Extract hour from time
df['hour'] = pd.to_datetime(df['time'], format='%H:%M:%S').dt.hour

# Sales metrics by hour
hourly_metrics = df.groupby('hour').agg({
    'price': 'sum',
    'quantity': 'sum', 
    'order_id': 'nunique'
}).round(2)
hourly_metrics.columns = ['Revenue', 'Pizzas_Sold', 'Orders']

# Identify peak hours
peak_revenue_hour = hourly_metrics['Revenue'].idxmax()
peak_orders_hour = hourly_metrics['Orders'].idxmax()
peak_quantity_hour = hourly_metrics['Pizzas_Sold'].idxmax()

print(f"Peak Performance Hours:")
print(f"  • Highest Revenue: {peak_revenue_hour}:00 (${hourly_metrics.loc[peak_revenue_hour, 'Revenue']:,.2f})")
print(f"  • Most Orders: {peak_orders_hour}:00 ({hourly_metrics.loc[peak_orders_hour, 'Orders']:.0f} orders)")
print(f"  • Most Pizzas: {peak_quantity_hour}:00 ({hourly_metrics.loc[peak_quantity_hour, 'Pizzas_Sold']:.0f} pizzas)")

# Operating hours analysis
operating_hours = hourly_metrics[hourly_metrics['Orders'] > 0].index
print(f"\nOperating Hours: {operating_hours.min()}:00 - {operating_hours.max()}:00")
print(f"Active Hours per Day: {len(operating_hours)}")

# Rush hours identification (top 25% of hours by orders)
rush_hour_threshold = hourly_metrics['Orders'].quantile(0.75)
rush_hours = hourly_metrics[hourly_metrics['Orders'] >= rush_hour_threshold].index
print(f"\nRush Hours (Top 25% busiest): {list(rush_hours)}")

# Quiet hours (bottom 25% of operating hours)
quiet_hour_threshold = hourly_metrics[hourly_metrics['Orders'] > 0]['Orders'].quantile(0.25)
quiet_hours = hourly_metrics[
    (hourly_metrics['Orders'] > 0) & 
    (hourly_metrics['Orders'] <= quiet_hour_threshold)
].index
print(f"Quiet Hours (Bottom 25%): {list(quiet_hours)}")

# Detailed hourly breakdown
print(f"\nComplete Hourly Breakdown:")
for hour, row in hourly_metrics.iterrows():
    if row['Orders'] > 0:
        status = ""
        if hour in rush_hours:
            status = " "
        elif hour in quiet_hours:
            status = " "
        
        print(f"  {hour:02d}:00 - Revenue: ${row['Revenue']:,.2f}, Orders: {row['Orders']:.0f}, Pizzas: {row['Pizzas_Sold']:.0f}{status}")

# Visualization
plt.figure(figsize=(18, 12))

plt.subplot(2, 3, 1)
hourly_metrics['Revenue'].plot(kind='bar', ax=plt.gca(), color='gold')
plt.axhline(hourly_metrics['Revenue'].mean(), color='red', linestyle='--', 
           label=f'Average: ${hourly_metrics["Revenue"].mean():.2f}')
plt.title('Revenue by Hour', fontsize=12, fontweight='bold')
plt.xlabel('Hour of Day')
plt.ylabel('Revenue ($)')
plt.legend()
plt.xticks(rotation=0)

plt.subplot(2, 3, 2)
hourly_metrics['Orders'].plot(kind='bar', ax=plt.gca(), color='skyblue')
plt.axhline(hourly_metrics['Orders'].mean(), color='red', linestyle='--',
           label=f'Average: {hourly_metrics["Orders"].mean():.1f}')
plt.title('Orders by Hour', fontsize=12, fontweight='bold')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.legend()
plt.xticks(rotation=0)

plt.subplot(2, 3, 3)
hourly_metrics['Pizzas_Sold'].plot(kind='bar', ax=plt.gca(), color='lightcoral')
plt.axhline(hourly_metrics['Pizzas_Sold'].mean(), color='red', linestyle='--',
           label=f'Average: {hourly_metrics["Pizzas_Sold"].mean():.1f}')
plt.title('Pizzas Sold by Hour', fontsize=12, fontweight='bold')
plt.xlabel('Hour of Day')
plt.ylabel('Pizzas Sold')
plt.legend()
plt.xticks(rotation=0)

plt.subplot(2, 3, 4)
hourly_metrics[['Revenue', 'Orders', 'Pizzas_Sold']].plot(ax=plt.gca())
plt.title('All Metrics by Hour (Normalized)', fontsize=12, fontweight='bold')
plt.xlabel('Hour of Day')
plt.ylabel('Relative Performance')
plt.legend()

plt.subplot(2, 3, 5)
# Heatmap of hourly performance
hourly_matrix = hourly_metrics[['Revenue', 'Orders', 'Pizzas_Sold']].T
sns.heatmap(hourly_matrix, annot=True, fmt='.0f', cmap='YlOrRd', ax=plt.gca())
plt.title('Hourly Performance Heatmap', fontsize=12, fontweight='bold')
plt.xlabel('Hour of Day')

plt.subplot(2, 3, 6)
# Average order value by hour
avg_order_value = hourly_metrics['Revenue'] / hourly_metrics['Orders']
avg_order_value.fillna(0).plot(kind='bar', ax=plt.gca(), color='purple')
plt.title('Average Order Value by Hour', fontsize=12, fontweight='bold')
plt.xlabel('Hour of Day')
plt.ylabel('Average Order Value ($)')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

### [7] Sales by day of the week - Which day has the most sales?

```python
print("\nWEEKLY SALES PATTERN ANALYSIS")
print("=" * 50)

# Weekly sales analysis
weekly_metrics = df.groupby('day_of_week').agg({
    'price': 'sum',
    'quantity': 'sum',
    'order_id': 'nunique'
}).round(2)
weekly_metrics.columns = ['Revenue', 'Pizzas_Sold', 'Orders']

# Reorder by standard week
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_metrics = weekly_metrics.reindex(day_order)

# Identify best and worst performing days
best_revenue_day = weekly_metrics['Revenue'].idxmax()
worst_revenue_day = weekly_metrics['Revenue'].idxmin()
best_orders_day = weekly_metrics['Orders'].idxmax()

print(f"BEST PERFORMING DAYS:")
print(f"  • Highest Revenue: {best_revenue_day} (${weekly_metrics.loc[best_revenue_day, 'Revenue']:,.2f})")
print(f"  • Most Orders: {best_orders_day} ({weekly_metrics.loc[best_orders_day, 'Orders']:.0f} orders)")

print(f"\nLOWEST PERFORMING DAY:")
print(f"  • Lowest Revenue: {worst_revenue_day} (${weekly_metrics.loc[worst_revenue_day, 'Revenue']:,.2f})")

# Calculate percentages and differences
total_weekly_revenue = weekly_metrics['Revenue'].sum()
total_weekly_orders = weekly_metrics['Orders'].sum()

print(f"\nDETAILED WEEKLY BREAKDOWN:")
for day, row in weekly_metrics.iterrows():
    revenue_pct = (row['Revenue'] / total_weekly_revenue) * 100
    orders_pct = (row['Orders'] / total_weekly_orders) * 100
    avg_order_value = row['Revenue'] / row['Orders'] if row['Orders'] > 0 else 0
    
    status = ""
    if day == best_revenue_day:
        status = " "
    elif day == worst_revenue_day:
        status = " "
    
    print(f"  {day}: ${row['Revenue']:,.2f} ({revenue_pct:.1f}%) | "
          f"{row['Orders']:.0f} orders ({orders_pct:.1f}%) | "
          f"${avg_order_value:.2f} avg{status}")

# Weekend vs Weekday analysis
weekend_days = ['Friday', 'Saturday', 'Sunday']
weekday_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday']

weekend_revenue = weekly_metrics.loc[weekend_days, 'Revenue'].sum()
weekday_revenue = weekly_metrics.loc[weekday_days, 'Revenue'].sum()

weekend_orders = weekly_metrics.loc[weekend_days, 'Orders'].sum()
weekday_orders = weekly_metrics.loc[weekday_days, 'Orders'].sum()

print(f"\nWEEKDAY vs WEEKEND COMPARISON:")
print(f"  Weekdays (Mon-Thu): ${weekday_revenue:,.2f} ({weekday_revenue/total_weekly_revenue*100:.1f}%)")
print(f"  Weekend (Fri-Sun): ${weekend_revenue:,.2f} ({weekend_revenue/total_weekly_revenue*100:.1f}%)")
print(f"  Weekend Premium: {((weekend_revenue/3) / (weekday_revenue/4) - 1)*100:.1f}% higher average daily revenue")

# Visualization
plt.figure(figsize=(18, 12))

plt.subplot(2, 3, 1)
weekly_metrics['Revenue'].plot(kind='bar', ax=plt.gca(), 
                              color=['lightcoral' if day == best_revenue_day else 'lightblue' for day in weekly_metrics.index])
plt.title('Revenue by Day of Week', fontsize=12, fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.subplot(2, 3, 2)
weekly_metrics['Orders'].plot(kind='bar', ax=plt.gca(),
                             color=['orange' if day == best_orders_day else 'lightgreen' for day in weekly_metrics.index])
plt.title('Orders by Day of Week', fontsize=12, fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)

plt.subplot(2, 3, 3)
weekly_metrics['Pizzas_Sold'].plot(kind='bar', ax=plt.gca(), color='gold')
plt.title('Pizzas Sold by Day of Week', fontsize=12, fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Pizzas Sold')
plt.xticks(rotation=45)

plt.subplot(2, 3, 4)
# Revenue percentage pie chart
revenue_pct = (weekly_metrics['Revenue'] / total_weekly_revenue) * 100
revenue_pct.plot(kind='pie', autopct='%1.1f%%', ax=plt.gca())
plt.title('Weekly Revenue Distribution', fontsize=12, fontweight='bold')
plt.ylabel('')

plt.subplot(2, 3, 5)
# Average order value by day
avg_order_values = weekly_metrics['Revenue'] / weekly_metrics['Orders']
avg_order_values.plot(kind='bar', ax=plt.gca(), color='purple')
plt.title('Average Order Value by Day', fontsize=12, fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Average Order Value ($)')
plt.xticks(rotation=45)

plt.subplot(2, 3, 6)
# Weekday vs Weekend comparison
comparison_data = pd.Series([weekday_revenue/4, weekend_revenue/3], 
                           index=['Weekday Avg', 'Weekend Avg'])
comparison_data.plot(kind='bar', ax=plt.gca(), color=['steelblue', 'darkorange'])
plt.title('Weekday vs Weekend Average Daily Revenue', fontsize=12, fontweight='bold')
plt.ylabel('Average Daily Revenue ($)')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

### [9] Sales made in each month

```python
print("\nMONTHLY SALES TRENDS ANALYSIS")
print("=" * 50)

# Monthly sales analysis
monthly_sales = df.groupby('month_name')['price'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

# Best and worst months
best_month = monthly_sales.idxmax()
worst_month = monthly_sales.idxmin()

print(f"Best Month: {best_month} (${monthly_sales.max():,.2f})")
print(f"Worst Month: {worst_month} (${monthly_sales.min():,.2f})")

# Calculate month-to-month change
monthly_change = monthly_sales.pct_change() * 100
print(f"\nNotable Changes:")
for month in monthly_change.dropna().index:
    change = monthly_change[month]
    if abs(change) > 10:  # Only show significant changes
        trend = "" if change > 0 else ""
        print(f"  {month}: {trend} {change:+.1f}% vs previous month")

# Visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
monthly_sales.plot(kind='line', marker='o', ax=plt.gca(), color='blue', linewidth=2)
plt.title('Monthly Revenue Trend', fontsize=12, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.subplot(1, 3, 2)
monthly_sales.plot(kind='bar', ax=plt.gca(), color='lightgreen')
plt.title('Monthly Revenue Comparison', fontsize=12, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
monthly_change.dropna().plot(kind='bar', ax=plt.gca(), 
                            color=['green' if x > 0 else 'red' for x in monthly_change.dropna()])
plt.title('Month-to-Month Change (%)', fontsize=12, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Change (%)')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### [10] Pizzas that aren't doing well

```python
print("\nUNDERPERFORMING PIZZAS ANALYSIS")
print("=" * 50)

# Identify underperforming pizzas
pizza_performance = df.groupby('name').agg({
    'quantity': 'sum',
    'price': 'sum',
}).round(2)
pizza_performance.columns = ['Total_Quantity', 'Total_Revenue']

# Define bottom 25% as underperformers
quantity_threshold = pizza_performance['Total_Quantity'].quantile(0.25)
underperformers = pizza_performance[pizza_performance['Total_Quantity'] <= quantity_threshold]
underperformers = underperformers.sort_values('Total_Quantity')

print(f"UNDERPERFORMING PIZZAS ({len(underperformers)} pizzas):")
print(f"Bottom 25% threshold: {quantity_threshold:.0f} pizzas or fewer\n")

print("Bottom 5 Performers:")
for i, (pizza, row) in enumerate(underperformers.head().iterrows(), 1):
    print(f"  {i}. {pizza}: {row['Total_Quantity']:,} sold, ${row['Total_Revenue']:,.2f} revenue")

# Visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
bottom_10 = underperformers.head(10)
pizza_names_short = [name[:20] + '...' if len(name) > 20 else name for name in bottom_10.index]
plt.barh(range(len(bottom_10)), bottom_10['Total_Quantity'], color='lightcoral')
plt.title('Bottom 10 Pizzas by Quantity', fontsize=12, fontweight='bold')
plt.xlabel('Quantity Sold')
plt.yticks(range(len(bottom_10)), pizza_names_short)
plt.gca().invert_yaxis()

plt.subplot(1, 3, 2)
plt.barh(range(len(bottom_10)), bottom_10['Total_Revenue'], color='lightblue')
plt.title('Bottom 10 Pizzas by Revenue', fontsize=12, fontweight='bold')
plt.xlabel('Revenue ($)')
plt.yticks(range(len(bottom_10)), pizza_names_short)
plt.gca().invert_yaxis()

plt.subplot(1, 3, 3)
# Performance distribution
pizza_performance['Total_Quantity'].hist(bins=20, alpha=0.7, color='lightgreen')
plt.axvline(quantity_threshold, color='red', linestyle='--', linewidth=2, 
           label=f'25th Percentile: {quantity_threshold:.0f}')
plt.title('Quantity Performance Distribution', fontsize=12, fontweight='bold')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Number of Pizzas')
plt.legend()

plt.tight_layout()
plt.show()
```
